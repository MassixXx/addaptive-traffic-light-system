```python
import numpy as np
from PIL import Image
from PIL import ImageFilter
%matplotlib inline
```

```python
import cv2
import matplotlib.pyplot as plt
 
im = cv2.imread('th.jpeg',0)
#im = cv2.equalizeHist(im)
plt.imshow(im)
plt.show()
edges = cv2.Canny(im,25,255)
plt.imshow(edges,cmap='gray')
plt.show()
```

```python
blur = cv2.bilateralFilter(edges, 15, 100, 100)
plt.imshow(blur,cmap='gray')
plt.show()
```

```python
threshold = 60
ret, thresh = cv2.threshold(blur, threshold, 255, cv2.THRESH_BINARY) #thresholding the frame
plt.imshow(thresh,cmap='gray')
plt.show()
```

```python
thresh = cv2.threshold(blur, 128, 255, cv2.THRESH_BINARY)[1]
rect=cv2.getStructuringElement(cv2.MORPH_RECT, (3, 3))
dilation = cv2.dilate(thresh,rect,iterations = 5)
erosion = cv2.erode(dilation, rect, iterations=4)
plt.imshow(erosion,cmap='gray')
plt.show()
```

```python
unique, frequency = np.unique(erosion,return_counts=True)
print(unique)
print(frequency)
```

```python
density = frequency[1] * 100 / (sum(frequency))
print(density)
```
